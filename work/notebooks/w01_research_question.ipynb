## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

Three real numbers from the starter dataset: total pages, the share
currently declining, and how many pages are both stale and still
getting real traffic - this last number is roughly the size of the
"worth reviewing" pool a hand rule would flag.

```python
import os, sys, subprocess

IN_COLAB = "google.colab" in sys.modules
REPO_URL = "https://github.com/Samme-creator/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"

if IN_COLAB:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
    os.chdir(REPO_DIR)

print("Working dir:", os.getcwd())
print("Data file exists:", os.path.exists("data/raw/content_refresh_anonymized.csv"))
```

```python
import pandas as pd
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
df["is_declining_label"] = df["trend_direction"].str.lower().eq("down").astype(int)

total_pages = df.shape[0]
declining_rate = df["is_declining_label"].mean()
stale_and_visible = ((df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)).sum()

print(f"Total pages: {total_pages:,}")
print(f"Declining rate: {declining_rate:.1%}")
print(f"Pages that are stale (180+ days) AND still visible (500+ impressions): {stale_and_visible:,}")
```
